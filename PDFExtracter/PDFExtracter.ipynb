```python
import fitz
from transformers import pipeline
```

```python
def extract_text_from_pdf(pdf_path):
    document = fitz.open(pdf_path)
    text = ""
    for page_num in range(document.page_count):
        page = document.load_page(page_num)
        text += page.get_text()
    return text
```

```python
def summarize_text(text, chunk_size=1024, max_summary_length=500):
    """Summarize text using a pre-trained transformer model."""
    summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
    text_chunks = [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]

    summary_chunks = []
    for chunk in text_chunks:
        input_length = len(chunk.split())
        max_length = min(max_summary_length, input_length)
        min_length = max(50, max_length // 2)  # min_length

        summary_chunk = summarizer(chunk, max_length=max_length, min_length=min_length, do_sample=False)[0]['summary_text']
        summary_chunks.append(summary_chunk)

    return " ".join(summary_chunks)
```

```python
def main(pdf_path):
    text = extract_text_from_pdf(pdf_path)

    summary = summarize_text(text, max_summary_length=500)

    return summary

if __name__ == "__main__":
    pdf_path = "Operations Management.pdf"
    summary = main(pdf_path)
    print("Summary:\n", summary)
```
